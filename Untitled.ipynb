```python
#neccessary library imports
import pandas as pd
import numpy as np
```

```python
#reading the data
data=pd.read_csv("insuranceFraud.csv")
```

```python
# having a look at top 5 rows of the data
data.head()
```

```python
# droping the columns that are not necessary for predection
columns_to_drop = ["policy_number","policy_bind_date","policy_state","insured_zip","incident_location","incident_date","incident_state","incident_city","insured_hobbies","auto_make","auto_model","auto_year"]
```

```python
#dropping unnecessary collumns
data = data.drop(columns_to_drop,axis =1)
```

```python
#in this dataset missing values are denoted by "?"
# so in the below step we are replacing the "?" with nan 
data=data.replace("?",np.nan)
```

```python
#checking for the categorical and numerical columns
data.info()
```

```python
data.isna().sum()
```

```python
#as the columns which have missing values are only the categorical data type 
#we use looping to fill with mode value
```

```python
for i in data.columns:
    try:
        mode= data[i].mode()[0]
        data[i]=data[i].fillna(mode)
    except:
        pass
```

```python
# encoding the data with our own custom values and converting the categorical values into numerical values
cat_df['policy_csl'] = cat_df['policy_csl'].map({'100/300':1,'250/500':2.5,'500/1000':5})
cat_df['insured_education_level']=cat_df['insured_education_level'].map({'MD':6, 'PhD':7, 'Associate':5, 'Masters':4, 'High School':2, 'College':3,
       'JD':1})
cat_df["incident_severity"]= cat_df["incident_severity"].map({'Major Damage':3, 'Minor Damage':2, 'Total Loss':4, 'Trivial Damage':1})
cat_df["insured_sex"] = cat_df["insured_sex"].map({'MALE':1, 'FEMALE':0})
cat_df["property_damage"] = cat_df["property_damage"].map({'YES':1, 'NO':0})
cat_df["police_report_available"] = cat_df["police_report_available"].map({'YES':1, 'NO':0})
cat_df['fraud_reported'] = cat_df['fraud_reported'].map({'Y':1, 'N':0})
```

```python
#auto encoding for the remaining categorical values

for col in cat_df.drop(columns = ['policy_csl', 'insured_sex', 'insured_education_level','insured_occupation',  'incident_severity','property_damage', 'police_report_available', 'fraud_reported']).columns:
    cat_df = pd.get_dummies(cat_df,columns = [col],prefix = [col],drop_first =True)
```
